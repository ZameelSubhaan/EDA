# ***Monthly Trends for City vs Resort Hotels***

```python
import plotly.express as px
# Filter to include only confirmed (non-canceled) bookings
confirmed_df = hotel_df[hotel_df['is_canceled'] == 0]

# Ensure months are in order
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
confirmed_df['arrival_date_month'] = pd.Categorical(confirmed_df['arrival_date_month'], categories=month_order, ordered=True)

# Group by hotel and month to calculate average ADR
adr_by_month = confirmed_df.groupby(['hotel', 'arrival_date_month'])['adr'].mean().reset_index()

# Plot
fig = px.line(
    adr_by_month,
    x='arrival_date_month',
    y='adr',
    color='hotel',
    title='Monthly ADR (Confirmed Bookings Only) for City vs Resort Hotel',
    markers=True,
    labels={'adr': 'Average Daily Rate', 'arrival_date_month': 'Month', 'hotel': 'Hotel Type'}
)

fig.update_layout(
    xaxis_title='Arrival Month',
    yaxis_title='Average Daily Rate (ADR)',
    legend_title='Hotel Type'
)

fig.show()
```

# **Monthly ADR (Confirmed Bookings Only) for City vs Resort Hotel**
**🔍Insight:**
**Resort Hotels:**

ADR peaks in August, showing a significant increase, indicating high demand — likely due to summer holidays.

ADR is low from November to February, showing it’s off-season during this time.

**City Hotels:**

ADR remains relatively stable, with slight increases in May–July.

The changes are less volatile, suggesting business/corporate bookings are a steady source of income.

**🎯 Possible Reasoning:**
Resort hotels are more dependent on leisure/tourist travel, which fluctuates with seasons.

City hotels attract more business or short-stay customers, which keeps their pricing more stable year-round.

**💡Actionable Suggestions to Improve Business:**
*Dynamic Off-Season Marketing for Resorts:

Offer discounts or bundled offers (like spa + stay or family activities) in low-ADR months (Nov–Feb) to attract more guests during the off-season.

*Leverage Peak Months with Tiered Pricing:

In August (peak month), introduce tiered pricing (standard, premium, deluxe) to maximize revenue from high-demand stays at resorts.

*Corporate & Event Partnerships for City Hotels:

Since city hotel ADR is stable, partner with companies or event organizers for long-term contracts to lock in business travelers and improve occupancy.

*Use Targeted Ads Based on Season:

Run targeted ads (Google, social media) promoting Resort Hotels during peak months and flash deals during off-seasons.

*Encourage Advance Bookings:

For both hotel types, offer early bird discounts for high-demand months like August, increasing cash flow and booking predictability.

*Introduce Loyalty Programs:

Develop loyalty rewards for guests who return during off-peak seasons — especially valuable for Resort Hotels, helping smooth the seasonal curve.

# ***Cancellation Rate by Market Segment (City vs Resort Hotels)***

```python
# Group by hotel and market segment to calculate cancellation rate
cancel_rate_df = (
    hotel_df.groupby(['hotel', 'market_segment'])['is_canceled']
    .mean()
    .reset_index()
    .rename(columns={'is_canceled': 'cancellation_rate'})
)

# Convert to percentage
cancel_rate_df['cancellation_rate'] = cancel_rate_df['cancellation_rate'] * 100

# Plot
fig = px.bar(
    cancel_rate_df,
    x='market_segment',
    y='cancellation_rate',
    color='hotel',
    barmode='group',
    title='Cancellation Rate by Market Segment (City vs Resort Hotel)',
    labels={'cancellation_rate': 'Cancellation Rate (%)', 'market_segment': 'Market Segment'}
)

fig.update_layout(
    xaxis_title='Market Segment',
    yaxis_title='Cancellation Rate (%)',
    legend_title='Hotel Type'
)

fig.show()
```

# **2. Cancellation Rate by Market Segment**
**🔍 Insight:**
Online Travel Agency (TA) segment has the highest cancellation rate for both City and Resort hotels.

Direct and Corporate bookings show much lower cancellation rates, indicating higher reliability.

City Hotels generally experience higher cancellation rates across segments compared to Resort Hotels.

**💡 Suggestions to Improve Cancellation Rates:**

Use stricter policies (like partial refund or non-refundable deals) for Online TA customers who are more likely to cancel.

*Incentivize Direct Bookings:

Offer small discounts or added perks (free breakfast, early check-in) to push more guests to book directly.

*Deposit Requirements for High-Risk Segments:

Require partial deposits from OTA and group bookings to discourage casual cancellations.

*Engage OTA Customers via Email/SMS:

Send reminders and pre-stay engagement emails to OTA bookers to lower the chance of no-shows or cancellations.

*Corporate Booking Contracts:

Strengthen contracts with corporate clients to lock in non-cancelable bookings for business stays.

*Use Machine Learning to Flag Risky Bookings:

Use historical booking data to predict high-risk cancellations, and flag such bookings for follow-up or stricter policy.

# ***Booking Lead Time vs Cancellation Rate (City vs Resort Hotel)***

```python
# Create bins for lead time
hotel_df['lead_time_bin'] = pd.cut(
    hotel_df['lead_time'],
    bins=[0, 7, 30, 60, 90, 120, 150, 180, 365],
    labels=['0-7', '8-30', '31-60', '61-90', '91-120', '121-150', '151-180', '181+']
)

# Group by hotel and lead time bin to calculate cancellation rate
lead_time_cancel_df = (
    hotel_df.groupby(['hotel', 'lead_time_bin'])['is_canceled']
    .mean()
    .reset_index()
    .rename(columns={'is_canceled': 'cancellation_rate'})
)

# Convert to percentage
lead_time_cancel_df['cancellation_rate'] = lead_time_cancel_df['cancellation_rate'] * 100

# Plot
fig = px.bar(
    lead_time_cancel_df,
    x='lead_time_bin',
    y='cancellation_rate',
    color='hotel',
    barmode='group',
    title='Cancellation Rate by Booking Lead Time (City vs Resort Hotel)',
    labels={'cancellation_rate': 'Cancellation Rate (%)', 'lead_time_bin': 'Booking Lead Time (Days)'}
)

fig.update_layout(
    xaxis_title='Lead Time Bin (Days Before Arrival)',
    yaxis_title='Cancellation Rate (%)',
    legend_title='Hotel Type'
)

fig.show()
```

# **3. Booking Lead Time vs Cancellation Rate**
🔍 Insight:
Longer lead times (more than 60 days) show significantly higher cancellation rates.

City Hotels tend to experience higher cancellations across almost all lead time bins compared to Resort Hotels.

The shortest lead times (0–7 days) have the lowest cancellation rates, especially in Resort Hotels — likely due to last-minute confirmed travel plans.

💡 Suggestions Based on This Insight:
*Encourage Shorter Booking Windows:

Offer discounts or limited-time offers for bookings made within 30 days to reduce cancellation risk.

*Stricter Policies for Long Lead Times:

For bookings with lead time > 60 days, enforce partial prepayment or cancellation fees.

*Overbooking Strategy:

City hotels can safely overbook a small percentage of long-lead-time bookings, as many will cancel.

*Advance Commitment Programs:

Give non-refundable discounted rates to customers who book early and are confident in their travel plans.

*Customer Behavior Scoring:

Flag bookings with long lead time + OTA segment as high-risk for cancellation and monitor closely.

*Target Local Last-Minute Travelers:

Use location-based ads to capture local weekend travelers who tend to book closer to their stay date and rarely cancel.

#**5** ***Bookings by Market Segment***

```python
# Grouping by country and hotel to analyze cancellations
country_cancel_df = (
    hotel_df.groupby(['country', 'hotel'])['is_canceled']
    .mean()
    .reset_index()
    .rename(columns={'is_canceled': 'cancellation_rate'})
)

# Filter out countries with very low bookings for better clarity
booking_counts = hotel_df['country'].value_counts()
popular_countries = booking_counts[booking_counts > 200].index  # Only include countries with >200 bookings
country_cancel_df = country_cancel_df[country_cancel_df['country'].isin(popular_countries)]

# Convert rate to percentage
country_cancel_df['cancellation_rate'] = country_cancel_df['cancellation_rate'] * 100

# Plot
fig = px.bar(
    country_cancel_df,
    x='country',
    y='cancellation_rate',
    color='hotel',
    barmode='group',
    title='Cancellation Rate by Country (City vs Resort Hotel)',
    labels={
        'cancellation_rate': 'Cancellation Rate (%)',
        'country': 'Country Code'
    }
)

fig.update_layout(
    xaxis_title='Country',
    yaxis_title='Cancellation Rate (%)',
    legend_title='Hotel Type',
    xaxis_tickangle=-45
)

fig.show()
```

✅ **Insights from the Analysis:**
Analyzing Bookings by Market Segment provides insight into the different customer types that contribute to bookings at both City and Resort hotels.

The market segment can reveal whether most bookings are coming from leisure travelers, business guests, or groups, which can influence pricing strategies, marketing efforts, and guest services.

# **Impact of Assigned Room Mismatch on Cancellations by Hotel Type**

```python
import plotly.express as px

# Create a mismatch flag
hotel_df['room_mismatch'] = hotel_df['reserved_room_type'] != hotel_df['assigned_room_type']

# Group by hotel and mismatch to calculate cancellation rate
room_mismatch_analysis = hotel_df.groupby(['hotel', 'room_mismatch']).agg(
    total_bookings=('is_canceled', 'count'),
    cancellations=('is_canceled', 'sum')
).reset_index()

room_mismatch_analysis['Cancellation Rate (%)'] = (
    room_mismatch_analysis['cancellations'] / room_mismatch_analysis['total_bookings'] * 100
)

# Replace True/False with readable labels
room_mismatch_analysis['Room Match Status'] = room_mismatch_analysis['room_mismatch'].map({
    True: 'Mismatch',
    False: 'Matched'
})

# Plot the graph
fig = px.bar(
    room_mismatch_analysis,
    x='Room Match Status',
    y='Cancellation Rate (%)',
    color='hotel',
    barmode='group',
    text='Cancellation Rate (%)',
    title='Impact of Room Assignment Mismatch on Cancellations',
    labels={'Room Match Status': 'Room Assignment Match'}
)

fig.update_traces(texttemplate='%{text:.2f}%', textposition='outside')
fig.update_layout(
    xaxis_title='Room Assignment Status',
    yaxis_title='Cancellation Rate (%)',
    legend_title='Hotel Type'
)

fig.show()
```

# **Length of Stay (LOS) vs ADR by Hotel Type for Weekdays and Weekends**

```python
import plotly.express as px

# Filter out invalid ADRs
valid_df = hotel_df[hotel_df['adr'] > 0]

# Calculate total nights and classify as weekend/weekday based on higher value
valid_df['total_nights'] = valid_df['stays_in_weekend_nights'] + valid_df['stays_in_week_nights']
valid_df = valid_df[valid_df['total_nights'] > 0]  # filter 0 nights just in case

# Classify stay type
def stay_type(row):
    if row['stays_in_weekend_nights'] > row['stays_in_week_nights']:
        return 'Weekend Stay'
    else:
        return 'Weekday Stay'

valid_df['Stay Type'] = valid_df.apply(stay_type, axis=1)

# Group by total_nights, hotel, stay_type and calculate mean ADR
los_adr = valid_df.groupby(['hotel', 'Stay Type', 'total_nights'])['adr'].mean().reset_index()

# Limit to reasonable length of stay for clarity
los_adr = los_adr[los_adr['total_nights'] <= 10]

# Plot
fig = px.line(
    los_adr,
    x='total_nights',
    y='adr',
    color='hotel',
    line_dash='Stay Type',
    markers=True,
    title='ADR vs Length of Stay (Weekday vs Weekend) by Hotel Type',
    labels={'adr': 'Average Daily Rate', 'total_nights': 'Length of Stay (Nights)'}
)

fig.update_traces(mode='lines+markers')
fig.update_layout(legend_title='Hotel / Stay Type')
fig.show()
```

# **Relationship Between Assigned Room Type Mismatch and Cancellation Rate**

```python
import plotly.express as px

# Create a column to detect mismatches
hotel_df['room_mismatch'] = hotel_df['reserved_room_type'] != hotel_df['assigned_room_type']

# Group by mismatch and hotel
mismatch_cancellation = hotel_df.groupby(['hotel', 'room_mismatch']).agg(
    total_bookings=('is_canceled', 'count'),
    cancellations=('is_canceled', 'sum')
).reset_index()

# Calculate cancellation rate
mismatch_cancellation['cancellation_rate'] = (
    mismatch_cancellation['cancellations'] / mismatch_cancellation['total_bookings']) * 100

# Plot
fig = px.bar(
    mismatch_cancellation,
    x='hotel',
    y='cancellation_rate',
    color='room_mismatch',
    barmode='group',
    title='Cancellation Rate vs Room Mismatch by Hotel Type',
    labels={'room_mismatch': 'Room Type Mismatch', 'cancellation_rate': 'Cancellation Rate (%)'}
)
fig.update_layout(xaxis_title="Hotel", yaxis_title="Cancellation Rate (%)")
fig.show()
```

Annoy guests.

Lead to cancellations or negative reviews.

Indicate overbooking, mismanagement, or inventory issues.

**Analyzing this helps improve:**

Guest satisfaction.

Internal room allocation systems.

Cancellation prevention strategies.
